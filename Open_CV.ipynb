# Image processing with OpenCV

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread('/Users/pranavbarot/Desktop/Okami.jpg')

plt.imshow(img)
```

## Invert blue and red colours, produce a grayscale image

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img2)
```

```python
gray_img = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)

plt.imshow(cv2.cvtColor(gray_img, cv2.COLOR_GRAY2RGB))
```

## Average colour being plotted from original (clearly blue)

```python
avg_color_per_row = np.average(img, axis=0)

avg_color = np.average(avg_color_per_row, axis=0)

avg_color = np.uint8(avg_color)
print(avg_color)
```

```python
avg_color_img = np.array([[avg_color]*100]*100, np.uint8)

plt.imshow(avg_color_img)
```

## Turning grayscale image to binary threshold image

```python
_, threshold_img = cv2.threshold(gray_img, 60, 255, cv2.THRESH_BINARY)

threshold_img = cv2.cvtColor(threshold_img, cv2.COLOR_GRAY2RGB)
plt.figure(figsize = (10,10))
plt.imshow(threshold_img)
```

```python
bounding = img
```

## Applying Gaussian Blur

```python
bounding_gray = cv2.cvtColor(bounding, cv2.COLOR_BGR2GRAY)
bounding_preprocessed = cv2.GaussianBlur(bounding_gray, (5,5), 0)

plt.imshow(cv2.cvtColor(bounding_preprocessed, cv2.COLOR_GRAY2RGB))
```

```python
bounding_preprocessed  = cv2.cvtColor(cv2.GaussianBlur(bounding, (7,7), 0), cv2.COLOR_BGR2GRAY)
```

```python
low_thresh = 100
high_thresh = 180

_, bounding_thresh = cv2.threshold(bounding_preprocessed, low_thresh, 255, cv2.THRESH_BINARY)
plt.imshow(cv2.cvtColor(bounding_thresh, cv2.COLOR_GRAY2RGB))
```

## Trying to detect boundaries, partially successful

```python
bounding_edges = cv2.Canny(bounding_preprocessed, threshold1=90, threshold2=110)
plt.imshow(cv2.cvtColor(bounding_edges, cv2.COLOR_GRAY2RGB))
```
