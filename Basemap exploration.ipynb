```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.basemap import Basemap
```

## Focus on Europe

```python
x1 = -20
x2 = 40
y1 = 30
y2 = 60

m = Basemap(resolution='l',projection='merc',\
            llcrnrlat=y1,urcrnrlat=y2,llcrnrlon=x1,\
            urcrnrlon=x2,lat_ts=(x1+x2)/2)
```

```python
lat = 41.3850639
lon = 2.1734034999999494
```

```python
plt.figure(figsize = (10,10))
m.drawcoastlines(linewidth = 3.0)
m.scatter(lon,lat,300)
m.drawcountries(linewidth = 2.5)
```

```python
x,y = m(lon, lat)
plt.figure(figsize = (10,10))
m.drawcoastlines(linewidth =3.0)
m.scatter(x,y,300)
m.drawcountries(linewidth = 2.5)
m.fillcontinents(color = 'orange')
```

## Specifying Barcelona's latitude and longitude, plotting the point

```python
plt.figure(figsize = (10,10))
m.drawcoastlines(linewidth =3.0)
m.scatter(x,y,300)
m.drawcountries(linewidth = 2.5)
m.drawmeridians(np.arange(0,360,10),labels=[False,True,True,False])
m.drawparallels(np.arange(-90,90,10),labels=[False,True,True,False])
m.drawmapscale(33,35,0,40,1000,barstyle='fancy')
m.fillcontinents(color = 'lightgray', lake_color = 'aqua', alpha =0.5)

plt.text(x+200000,y-500000, 'Barcelona', color = 'red', size = 14)

plt.show()
```

## Topography depiction

```python
plt.figure(figsize = (10,10))
m.drawcoastlines(linewidth =2.0)
m.scatter(x,y,300)
m.drawcountries(linewidth = 1.5)
m.fillcontinents(color='lightgray',lake_color='aqua',alpha=0.5)
m.etopo()
m.scatter(x,y,300, zorder = 10)
```

```python
plt.figure(figsize = (10,10))
plt.text(x+200000,y-500000, 'Barcelona', color = 'red', size = 14)
m.drawcoastlines(linewidth =2.0)
m.scatter(x,y,300)
m.drawcountries(linewidth = 1.5)
m.fillcontinents(color='lightgray',lake_color='aqua',alpha=0.5)
m.shadedrelief()
m.scatter(x,y,zorder=8)
```
