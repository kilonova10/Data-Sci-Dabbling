```python
import spacy
import pandas as pd
import os
```

# Load data

```python
# Get you own transcripts here ...
transcripts_df = pd.read_json(os.path.join("..","..","TalkIQ","CNN-models","data","transcripts","broadly-10k.json"), lines=True)
```

```python
# Crop for demo purposes and speed
transcripts_df = transcripts_df[0:10]
```

```python
# Convert to unicode for scaCy
transcripts_df['raw_text'] = transcripts_df['taste'].apply(lambda s: ' '.join([w['raw'] for w in s if w['raw'] != '[noise]']).decode('utf8'))
```

```python
# Preview
transcripts_df[0:1]
```

```python
nlp = spacy.load('en')
```

```python
text = ' '.join(transcripts_df['raw_text'].tolist())
```

The nlp instance is used to produce **Doc** objects. You'll then use the Doc object to access linguistic annotations to help you with whatever text processing task you're trying to do.

```python
%%time
parsed_text = nlp(text)
```

```python
for num, sentence in enumerate(parsed_text.sents):
    print 'Sentence {}:'.format(num + 1)
    print sentence
    print ''
```

```python
for num, entity in enumerate(parsed_text.ents):
    print 'Entity {}:'.format(num + 1), entity, '-', entity.label_
    print ''
```

```python
token_text = [token.orth_ for token in parsed_text]
token_pos = [token.pos_ for token in parsed_text]

pos_df = pd.DataFrame(zip(token_text, token_pos),
             columns=['token_text', 'part_of_speech'])
```
